```python
from PIL import Image, ImageCms
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = Image.open('GitHub\Lego-Vincent-Van-Gogh-Starry-Night\Final 116x92 pixel Recoloured Starry Night.png')
rgbImg = img.convert('RGB')
plt.figure(figsize=(11,9))
plt.imshow(rgbImg)
```

```python
imgCol = Image.open('LegoColours.png')
plt.figure(figsize=(11,9))
plt.imshow(imgCol)
```

```python
count = np.zeros([22]);

for i in range(rgbImg.size[0]):
    for j in range(rgbImg.size[1]):
        match = False;
        for k in range(len(count)):
            if rgbImg.getpixel((i,j)) == imgCol.getpixel((0,k)):
                count[k] += 1;  
                match = True;
        if match != 1:
            print(rgbImg.getpixel((i,j)), " ", i, " ", j)
```

```python
for val in count: print(val);
```
