```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
%matplotlib inline
```

```python
img = Image.open('StarryNightPixelated.png')
plt.figure(figsize=(11,9))
plt.imshow(img)
```

```python
colours = np.loadtxt("colourList.csv", delimiter = ',', dtype = int)
# print(colours)
```

```python
ReColour = np.zeros_like(img)

for i in tqdm(range(img.size[0])):
    for j in range(img.size[1]):
        ReColour[j,i] = colours[np.argmin(np.sqrt(np.sum((img.getpixel((i,j)) - colours)**2, axis=1)))]
```

```python
plt.figure(figsize=(11,9))
plt.imshow(ReColour)
```

```python
imgReColour = Image.fromarray(ReColour, mode = 'RGB')
```

```python
imgReColour.save("StarryNightRecoloured.png")
```
